```python
import requests
import pandas as pd
import flair
import re
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
import yfinance as yf
import matplotlib.pyplot as plt
import twint


config = twint.Config()
config.Username = 'Microsoft'
config.Lang = "en"
config.Since = '2006-12-01 01:00:00'
config.Until = '2021-07-06 01:00:00'
config.Store_csv = True
config.Count = True 
config.Retweets = True
config.Limit = 100000
config.Output = "msft_tweets.csv"
twint.run.Search(config)
```

```python
ugly_df = pd.read_csv('msft_tweets.csv', parse_dates=['created_at'])
df = pd.DataFrame()

df['Date'] = ugly_df.date
df['tweet'] = ugly_df.tweet


df.head()
```

```python
new_list = []
for tweet in df['tweet']:
    whitespace = re.compile(r"\s+")
    web_address = re.compile(r"(?i)http(s):\/\/[a-z0-9.~_\-\/]+")
    tesla = re.compile(r"(?i)@Tesla(?=\b)")
    user = re.compile(r"(?i)@[a-z0-9_]+")

# we then use the sub method to replace anything matching
    tweet = whitespace.sub(' ', tweet)
    tweet = web_address.sub('', tweet)
    tweet = tesla.sub('Tesla', tweet)
    tweet = user.sub('', tweet)
    new_list.append(tweet)

text_df = pd.DataFrame(new_list, columns=['text'])
text_df.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))


df['tweet'] = text_df['text']

```

```python
#Combining all the tweets posted on a single date
df['tweet'] = df.groupby(['Date'])['tweet'].transform(lambda x: ' '.join(x))
```

```python
from tqdm import tqdm
def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase

def clean(text):
    text = str(text)
    text = text.lower()
    text = re.sub(r'http\S+', ' ', text)
    text = re.sub(r'pic.twitter\S+', ' ', text)
    text = decontracted(text)
    text = re.sub(r'\(([^)]+)\)', " ", text)
    text = text.replace('etmarkets', ' ').replace('marketupdates', ' ').replace('newsalert', ' ').replace('ndtv', ' ').replace('moneycontrol', ' ').replace('here is why', ' ')
    text = text.replace('marketsupdate', ' ').replace('biznews', ' ').replace('click here', ' ').replace('live updates', ' ').replace('et now', ' ')
    text = re.sub(r'[^a-zA-Z ]+', ' ', text)
    text = re.sub(r' \w{1,2}_', ' ', text)
    text = re.sub('\s+',' ', text)
    return text

for i in tqdm(df.itertuples()):
    df.at[i[0], 'tweet'] = clean(i[2])

```
